### Adding spelling mistakes

```python
dfSpell = dfClean.sample(frac=0.1)
```

```python
len(dfSpell)
```

```python
columns = ["HospitalName", "Address1","City","State", "CountyName","HospitalType"]
```

```python
def modifyValue(value):
    wordList = list(value)
    randomPosition = random.randint(0, len(value) - 1)
    currentChar = wordList[randomPosition]
    
    lowercase = set(string.ascii_lowercase)
    uppercase = set(string.ascii_uppercase)
    digits = set(string.digits)
    
    while currentChar.isspace() is True:
        randomPosition = random.randint(0, len(value) - 1)
        currentChar = wordList[randomPosition]
    
    if currentChar.isalpha():
        if currentChar.islower():
            char = random.choice(list(lowercase - set(currentChar)))
        else:
            char = random.choice(list(uppercase - set(currentChar)))
    else:
        char = random.choice(list(digits - set(currentChar)))
    
    wordList[randomPosition] = str(char)
    return ''.join(wordList)
```

```python
for index, row in dfSpell.iterrows():
    column = random.choice(columns)
    value = row[column]
    modifiedValue = modifyValue(value)
    dfSpell.at[index, column] = modifiedValue
    dfSpell.at[index, 'label'] = 0
```

```python
df = pd.concat([df, dfSpell])
```
